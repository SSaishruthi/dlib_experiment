```python
import dlib
import cv2
from IPython.display import Image, display
```

# CNN DLIB

```python
image_cnn = cv2.imread('trial.jpg')
```

```python
cnn_face_detector = dlib.cnn_face_detection_model_v1('mmod_human_face_detector.dat')
```

```python
faces_cnn = cnn_face_detector(image_cnn, 1)
```

```python
for face in faces_cnn:
    x = face.rect.left()
    y = face.rect.top()
    w = face.rect.right() - x
    h = face.rect.bottom() - y

     # draw box over face
    cv2.rectangle(image_cnn, (x,y), (x+w,y+h), (0,0,255), 2)
```

```python
img_height, img_width = image_cnn.shape[:2]
```

```python
image_cnn = cv2.putText(image_cnn, "CNN", (img_width-50,40), cv2.FONT_HERSHEY_SIMPLEX, 0.5,
                (0,0,255), 2)
```

```python
cv2.imwrite("trial_cnn.jpg", image_cnn)
```

```python
display(Image(filename='trial_cnn.jpg', embed=True))
```

# HOG Dlib

```python
#HOG
image_hog = cv2.imread('trial.jpg')
hog_face_detector = dlib.get_frontal_face_detector()
faces_hog = hog_face_detector(image_hog, 1)
for face in faces_hog:
    x = face.left()
    y = face.top()
    w = face.right() - x
    h = face.bottom() - y

    # draw box over face
    cv2.rectangle(image_hog, (x,y), (x+w,y+h), (0,255,0), 2)
```

```python
img_height, img_width = image_hog.shape[:2]
```

```python
image_hog = cv2.putText(image_hog, "HOG", (img_width-50,20), cv2.FONT_HERSHEY_SIMPLEX, 0.5,
                (0,255,0), 2)
```

```python
cv2.imwrite("trial_hog.jpg", image_hog)
```

```python
display(Image(filename='trial_hog.jpg', embed=True))
```
